### Defining the convolution functions

```python
def standard_conv(x, h):
    
    Nx = len(x)
    Nh = len(h)
    X = np.hstack((x, np.zeros((Nx,), float)))
    H = np.hstack((h, np.zeros((Nh,), float)))
    
    
    Ny = Nx + Nh - 1
    y = np.zeros((Ny,), float)

    for p in range(Ny):
        for q in range(Nh):
            if (p - q + 1 > 0):
                y[p] += H[q] * X[ p - q]
                
    return y


def alternate_conv(x, h):
    
    Nx = len(x)
    Nh = len(h)
    X = np.hstack((x, np.zeros((Nx,), float)))
    H = np.hstack((h, np.zeros((Nh,), float)))
    
    
    Ny = Nx + Nh - 1
    y = np.zeros((Ny,), float)

    for p in range(Nx):
        for q in range(Nh):
            y[ p + q ] += X[p] * H[q]
                
    return y


def built_in_conv(x, h):
    
    bi = np.convolve(x, h)
    return bi
```

#### Calling the functions

```python
s_conv = standard_conv(x, h)
```

```python
a_conv = alternate_conv(x, h)
```

```python
bi_conv = built_in_conv(x, h)
```

#### Plotting

```python
plt.plot(s_conv)
plt.title('Standard Convoluted output y')
plt.xlabel('Ny')
plt.ylabel('Magnitude')
plt.show()
```

```python
plt.plot(a_conv)
plt.title('Alternate Convoluted output y')
plt.xlabel('Ny')
plt.ylabel('Magnitude')
plt.show()
```
